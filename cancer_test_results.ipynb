# Cancer Test Results

```python
# load dataset
import numpy as np
import pandas as pd
data = pd.read_csv('cancer_test_data.csv')
data.head()
```

```python
# number of patients
data.shape[0]
```

```python
# number of patients with cancer
data[data['has_cancer'] == True].count()
```

```python
# number of patients without cancer
data[data['has_cancer'] == False].count()
```

```python
# proportion of patients with cancer
data.has_cancer.mean()
```

```python
# proportion of patients without cancer
1-data.has_cancer.mean()
```

```python
data.query('has_cancer')
```

```python
data.query('has_cancer == False')
```

```python
# proportion of patients with cancer who test positive
(data.query('has_cancer')['test_result'] == 'Positive').mean()
```

```python
# proportion of patients with cancer who test negative
(data.query('has_cancer')['test_result'] == 'Negative').mean()
```

```python
# proportion of patients without cancer who test positive
(data.query('has_cancer == False')['test_result'] == 'Positive').mean()
```

```python
# proportion of patients without cancer who test negative
(data.query('has_cancer == False')['test_result'] == 'Negative').mean()
```
